```python
with open("mr_IN.dic") as f:
    dc = [line.rstrip('\n') for line in f]   

with open("DocumentList.xml") as f:
    ac = list(f)
```

```python
newlist = list()
for i in ac:
    newlist.append(re.findall(r'"([^"]*)"', i))
```

```python
wrong = list()
right = list()
for i in newlist:
    wrong.append(i[0])
    right.append(i[1])
    
w = set(wrong)
r = set(right)
d = set(dc)
```

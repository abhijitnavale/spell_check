```python
import hunspell

spellchecker = hunspell.HunSpell(
    "/tmp/spell_check/mymarathi/dicts/mr_IN.dic",
    "/tmp/spell_check/mymarathi/dicts/mr_IN.aff",
)
```

```python
from bs4 import BeautifulSoup
import requests, re
```

```python
url = "http://pashanbhed.blogspot.com/2020/01/blog-post.html"
```

```python
def mycheck(myword):
    if spellchecker.spell(myword) is False:
        try:
            print(myword, spellchecker.suggest(myword)[0])
        except:
            pass
```

```python
headers = requests.utils.default_headers()
req = requests.get(url, headers)
soup = BeautifulSoup(req.content, "html.parser")
text = soup.get_text()

p = re.compile(r"[^\u0900-\u097F\n]")
for line in text.splitlines():
    cleaned = p.sub(" ", line)
    if cleaned.strip():
        for i in cleaned.split():
            mycheck(i)
```
